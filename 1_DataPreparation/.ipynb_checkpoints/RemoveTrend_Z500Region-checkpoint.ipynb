```python
def detrend_constant(dataset, var_name_xarray, trend):
    dataset = dataset.isel(time=(pd.to_datetime(dataset.time).year >= 1981))
    lat = dataset.lat.values
    lon = dataset.lon.values

    for lati in range(len(lat)):
        for loni in range(len(lon)):
            series = dataset.sel(lat=lat[lati], lon=lon[loni])[f'{var_name_xarray}_anomalies']
            X = np.arange(len(series))
            rect_line = X * trend
            dataset[f'{var_name_xarray}_anomalies'].loc[dict(lat=lat[lati], lon=lon[loni])] -= rect_line

    return dataset
```

```python
def weighted_average_2d_array(data_array):
    weights = np.cos(np.deg2rad(data_array.lat))
    weights.name = "weights"
    data_weighted = data_array.weighted(weights)
    weighted_mean = data_weighted.mean(("lon", "lat"))

    return weighted_mean
```

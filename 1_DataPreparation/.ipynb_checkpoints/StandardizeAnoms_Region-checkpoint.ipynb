```python
def get_climatology_std(dataset,var_name_xarray,path_save_climatology):
    dataset_clima = dataset.isel(time = (pd.to_datetime(dataset.time).year>=1981)&\
                                 (pd.to_datetime(dataset.time).year<=2015))
    # return dataset_clima
    dataset_clima = dataset_clima.isel(time = ~((pd.to_datetime(dataset_clima.time).day == 29)&((pd.to_datetime(dataset_clima.time).month == 2))))
    
    doy = pd.to_datetime(dataset_clima.time).day_of_year
    climatology = []
    for i in range(1,366):
        array_temp = dataset_clima.isel(time = doy == i)[var_name_xarray]
        std = np.nanstd(array_temp,axis=0)
        std[std==0] = np.nan
        climatology.append(std)
    attrs = dataset[var_name_xarray].attrs
    attrs['File Author'] = 'Jhayron S. PÃ©rez-Carrasquilla'
    
    climatology = xr.Dataset({
                 f'{var_name_xarray}_climatology_std': (['day_of_year','lat','lon'], np.array(climatology)),
                },
                 coords =
                {'day_of_year': (['day_of_year'], np.arange(1,366)),
                 'lat' : (['lat'], dataset.lat.values),
                 'lon' : (['lon'], dataset.lon.values)
                },
                attrs = attrs)
    
    climatology = climatology.transpose('day_of_year','lat','lon')
    climatology.to_netcdf(path_save_climatology)
    return climatology
```

```python
def standardize_anomalies(anomalies,var_name_xarray,climatology_std,path_save_anomalies):
    std_anomalies = copy.deepcopy(anomalies)
    for day in range(1,367):
        # print(day) 
        if day == 366:
            std_anomalies[var_name_xarray][{'time':(pd.to_datetime(anomalies.time).day_of_year == day)}] = \
                (anomalies[var_name_xarray].isel(time = (pd.to_datetime(anomalies.time).day_of_year == day)) \
                / climatology_std[f'{var_name_xarray}_climatology_std'.replace('anomalies_','')].sel(day_of_year = day-1))
        else:
            std_anomalies[var_name_xarray][{'time':(pd.to_datetime(anomalies.time).day_of_year == day)}] = \
                (anomalies[var_name_xarray].isel(time = (pd.to_datetime(anomalies.time).day_of_year == day)) \
                / climatology_std[f'{var_name_xarray}_climatology_std'.replace('anomalies_','')].sel(day_of_year = day))
    # std_anomalies = std_anomalies.rename({var_name_xarray:f'{var_name_xarray}_anomalies'})
    std_anomalies.to_netcdf(path_save_anomalies)
    return std_anomalies
```

```python
def standardize_anomalies_region(anomalies,var_name_xarray,climatology_std,path_save_anomalies):
    std_anomalies = copy.deepcopy(anomalies)
    lat0 = 20; lat1 = 80; lon0 = 180; lon1 = 330
    for day in range(1,367):
        # print(day) 
        if day == 366:
            std_region = climatology_std[f'{var_name_xarray}_climatology_std'.replace('anomalies_','')].sel(day_of_year = day-1)
            std_region = std_region.mean().values[()]
            std_anomalies[var_name_xarray][{'time':(pd.to_datetime(anomalies.time).day_of_year == day)}] = \
                (anomalies[var_name_xarray].isel(time = (pd.to_datetime(anomalies.time).day_of_year == day)) \
                / std_region)
        else:
            std_region = climatology_std[f'{var_name_xarray}_climatology_std'.replace('anomalies_','')].sel(day_of_year = day)
            std_region = std_region.mean().values[()]
            std_anomalies[var_name_xarray][{'time':(pd.to_datetime(anomalies.time).day_of_year == day)}] = \
                (anomalies[var_name_xarray].isel(time = (pd.to_datetime(anomalies.time).day_of_year == day)) \
                / std_region)
    # std_anomalies = std_anomalies.rename({var_name_xarray:f'{var_name_xarray}_anomalies'})
    std_anomalies.to_netcdf(path_save_anomalies)
    return std_anomalies
```

```python
path_figures_seasonal = 'figures_random_forest/seasonal/'
```

```python
for var in names_vars:
    weeks = np.arange(1,9)
    
    accs_winter = []
    accs_spring = []
    accs_summer = []
    accs_fall = []

    for week_out in weeks:
        results_temp = pd.read_csv(f'{path_results}results_{var}_week{week_out}.csv',
                                  index_col=0,parse_dates=True)

        names_seasons = ['Winter','Spring','Summer','Fall']

        df_winter = results_temp[(results_temp.index.month==12)|\
                    (results_temp.index.month==1)|\
                    (results_temp.index.month==2)]
        df_spring = results_temp[(results_temp.index.month==3)|\
                    (results_temp.index.month==4)|\
                    (results_temp.index.month==5)]
        df_summer = results_temp[(results_temp.index.month==6)|\
                    (results_temp.index.month==7)|\
                    (results_temp.index.month==8)]
        df_fall = results_temp[(results_temp.index.month==9)|\
                    (results_temp.index.month==10)|\
                    (results_temp.index.month==11)]

        acc_winter = accuracy_score(df_winter['y_test'],df_winter['y_predicted'])
        acc_spring = accuracy_score(df_spring['y_test'],df_spring['y_predicted'])
        acc_summer = accuracy_score(df_summer['y_test'],df_summer['y_predicted'])
        acc_fall = accuracy_score(df_fall['y_test'],df_fall['y_predicted'])

        accs_winter.append(acc_winter)
        accs_spring.append(acc_spring)
        accs_summer.append(acc_summer)
        accs_fall.append(acc_fall)
        
    plt.plot(weeks[1:],np.array(accs_winter)[1:] * 100,color='blue',label='Winter')
    plt.plot(weeks[1:],np.array(accs_spring)[1:] * 100,color='green',label='Spring')
    plt.plot(weeks[1:],np.array(accs_summer)[1:] * 100,color='red',label='Summer')
    plt.plot(weeks[1:],np.array(accs_fall)[1:] * 100,color='darkorange',label='Fall')

    plt.legend()
    plt.ylim(13,42)
    plt.grid(alpha=0.3)
    plt.ylabel('Accuracy [%]')
    # plt.xlabel('Duration (weeks)')
    plt.title(f'Seasonal changes in accuracy: {var}',loc='left')
    
    plt.savefig(f'{path_figures_seasonal}seasonal_acc_{var}.png',bbox_inches='tight')
    plt.close('all')
```
